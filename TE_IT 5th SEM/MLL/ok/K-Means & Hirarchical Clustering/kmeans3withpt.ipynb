```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Given points
points = np.array([[0.1, 0.6], [0.15, 0.71], [0.08, 0.9], [0.16, 0.85], [0.2, 0.3], [0.25, 0.5], [0.24, 0.1], [0.3, 0.2]])

# Initial centroids
initial_centroids = np.array([[0.1, 0.6], [0.3, 0.2]])

# Perform K-Means clustering
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1, random_state=0)
kmeans.fit(points)

# Get cluster labels and centroids
cluster_labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Plotting the clusters and centroids
plt.figure(figsize=(8, 6))

# Plot each point with its assigned color
for i in range(len(points)):
    color = 'r' if cluster_labels[i] == 0 else 'b'
    plt.scatter(points[i, 0], points[i, 1], c=color, marker='o')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='g', marker='X', s=100, label='Centroids')

# Highlight P6
p6_cluster = kmeans.predict([[0.25, 0.5]])
plt.scatter(0.25, 0.5, c='purple', marker='s', s=100, label='P6')

plt.title('K-Means Clustering')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
