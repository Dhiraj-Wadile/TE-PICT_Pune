```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("cities_r2.csv")
```

```python
df
```

```python
df.columns
```

```python
x=df[['effective_literacy_rate_total']]
y=df[['dist_code']]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x,y)
```

```python
from sklearn.metrics import silhouette_score
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
sil=[]
for k in range(2,16):
    km = KMeans(n_clusters = k)
    l = km.fit_predict(x)
    score = silhouette_score(x,l)
    sil.append(score)
    
print(sil)    
    
    
```

```python
plt.bar(range(2,16),sil)
```

```python
agl = AgglomerativeClustering(n_clusters=2)
alabel = agl.fit_predict(x)
```

```python
plt.scatter(x,y,c=alabel)
```
