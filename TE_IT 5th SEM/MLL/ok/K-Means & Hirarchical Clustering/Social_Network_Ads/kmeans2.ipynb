```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv(r"D:\ml dataset\LP-1 Dataset\K-Means & Hirarchical Clustering\Social_Network_Ads.csv")
```

```python
df
```

```python
x=df[['EstimatedSalary']]
y=df[['User ID']]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x['EstimatedSalary'],y['User ID'])
```

```python
from sklearn.cluster import KMeans
```

```python
sse = []
for k in range(1,15):
    km = KMeans(n_clusters=k)
    km.fit_predict(x)
    sse.append(km.inertia_)
    
print(sse)    
```

```python
plt.xticks(range(1,15))
plt.plot(range(1,15),sse,marker='.')
```

```python
km = KMeans(n_clusters=4,random_state=0)
labels = km.fit_predict(x)
plt.scatter(x['EstimatedSalary'],y['User ID'],c=labels)
```
