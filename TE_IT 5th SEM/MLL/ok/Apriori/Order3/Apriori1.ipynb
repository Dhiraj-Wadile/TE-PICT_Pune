```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
df = pd.read_csv(r"D:\ml dataset\LP-1 Dataset\Apriori\Oder3.csv")
```

```python
df
```

```python
te = TransactionEncoder()
```

```python
x = te.fit_transform(df)
```

```python
x
```

```python
data = pd.DataFrame(x,columns=te.columns_)
```

```python
len(data)
```

```python
data.head()
```

```python
data
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
freq_itemset = apriori(data,min_support=0.00001,use_colnames=True)
```

```python
print(freq_itemset)
```

```python
len(te.columns_)
```

```python
#Find Rules

rules = association_rules(freq_itemset,metric = 'confidence',min_threshold=0.25)
```

```python
rules = rules[['antecedents','consequents','antecedent support','support','confidence']]
```
