```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
df = pd.read_csv(r"D:\ml dataset\LP-1 Dataset\Apriori\Order2.csv")
```

```python
tran = []
for row in df.values:
    tran.append([item.strip() for item in row if pd.notna(item)])
```

```python
tran
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
x = te.fit_transform(tran)
df = pd.DataFrame(x,columns = te.columns_)
```

```python
df.head()
```

```python
freq_item = apriori(df,min_support=0.01,use_colnames=True)
```

```python
freq_item
```

```python
rules = association_rules(freq_item,min_threshold=0.01,metric='confidence')
```

```python
rules.head(5)
```

```python
for i,row in rules.iterrows():
    antecedents = ', '.join(row['antecedents'])
    consequents = ', '.join(row['consequents'])
    support = row['support']
    print(f"rule is {i+1}")
    print(f"antecedents : {antecedents}")
    print(f"consequents : {consequents}")
    print(f"support     : {support:.4f}")
    print("="*30)
```
